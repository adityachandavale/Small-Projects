```python
import pandas as pd

url = "https://docs.google.com/spreadsheets/d/1M2R2Krbdlhhb9-pGEOIdul-0d7H2YzBZpv8NzgUKkzE/export?format=csv&gid=266531591"
df = pd.read_csv(url)

df['What options of Investment are you aware of?'] = df['What options of Investment are you aware of?'].fillna("na")

cnt_savings = 0
cnt_bonds = 0
cnt_fixed = 0
cnt_mutual = 0
cnt_crypto = 0
cnt_stock = 0
cnt_deb = 0
cnt_real = 0
cnt_gld = 0
cnt_ins = 0

for i in range(len(df['What options of Investment are you aware of?'])):
    for j in range(len(df['What options of Investment are you aware of?'][i].split(','))):
        if df['What options of Investment are you aware of?'][i].split(',')[j] == 'Savings account':
            cnt_savings += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == 'Bonds' or df['What options of Investment are you aware of?'][i].split(',')[j] == ' Bonds':
            cnt_bonds += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == ' Fixed Deposits' or df['What options of Investment are you aware of?'][i].split(',')[j] == 'Fixed Deposits':
            cnt_fixed += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == ' Mutual Funds' or df['What options of Investment are you aware of?'][i].split(',')[j] == 'Mutual Funds':
            cnt_mutual += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == ' Crypto currencies' or df['What options of Investment are you aware of?'][i].split(',')[j] == 'Crypto currencies':
            cnt_crypto += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == ' Stock market' or df['What options of Investment are you aware of?'][i].split(',')[j] == 'Stock market':
            cnt_stock += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == ' Debentures' or df['What options of Investment are you aware of?'][i].split(',')[j] == 'Debentures':
            cnt_deb += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == ' Real estate' or df['What options of Investment are you aware of?'][i].split(',')[j] == 'Real estate':
            cnt_real += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == ' Gold & Silver' or df['What options of Investment are you aware of?'][i].split(',')[j] == 'Gold & Silver':
            cnt_gld += 1
        
        
        if df['What options of Investment are you aware of?'][i].split(',')[j] == ' Insurance' or df['What options of Investment are you aware of?'][i].split(',')[j] == 'Insurance':
            cnt_ins += 1
        
l = {"Savings account":cnt_savings, "Bonds":cnt_bonds, "Fixed Deposits":cnt_fixed, "Mutual Funds":cnt_mutual, "Crypto Currencies":cnt_crypto, "Stock Market":cnt_stock, "Debentures":cnt_deb, "Real Estate":cnt_real, "Glod & Silver":cnt_gld, "Insurance":cnt_ins}

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,10))
plt.bar(l.keys(),l.values())
plt.title("What options of Investment are you aware of?")
plt.show()
```
